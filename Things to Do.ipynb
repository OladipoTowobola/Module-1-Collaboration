```python
## 4.1 I used comparison operators to determine the relationship between your guess and the secret number.
secret = 7
guess = 3

if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')
```

```python
## 4.2  I used Boolean logic to match a combination of characteristics (size and color) to specific items.
# Change these values to test different outcomes
small = True
green = False

if small:
    if green:
        print("pea")
    else:
        print("cherry")
else:
    if green:
        print("watermelon")
    else:
        print("pumpkin")
```

```python
##6.2 Assigning the value 7 to the variable guess_me, and the value 1 to the variable number
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
    number += 1
```

```python
## 6.3 The range(10) function generates a sequence of numbers from 0 to 9.
guess_me = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
