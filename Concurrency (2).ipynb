13.1 Write the current date as a string to the text file today.txt

```python
from datetime import datetime, date, timedelta
current_date = date.today()
with open('today.txt', 'w') as f:
    f.write(current_date.isoformat())
```

13.2 Read the text file today.txt into the string today_string

```python
with open("today.txt", "r") as file:
    today_string = file.read()

print("Today's date string from file:", today_string)
```

13.3 Parse the date from today_string

```python
parsed_date = datetime.strptime(today_string, "%Y-%m-%d").date()
print("Parsed date object:", parsed_date)
```

15.1 Use multiprocessing to create three separate processes. 
Make each one wait a random number of seconds between zero and one, print the current time, and then exit.


```python
import multiprocessing
import time
import random
from datetime import datetime

def print_time_after_delay():
    """Waits for a random time and prints the current timestamp."""
    # Generate a random float between 0 and 1
    delay = random.random()
    time.sleep(delay)
    
    # Get current time formatted as HH:MM:SS.ms
    current_time = datetime.now().strftime("%H:%M:%S.%f")[:-3]
    
    print(f"Process {multiprocessing.current_process().name} waited {delay:.2f}s - Current Time: {current_time}")

if __name__ == "__main__":
    processes = []

    # Create and start three separate processes
    for i in range(3):
        p = multiprocessing.Process(target=print_time_after_delay, name=f"Worker-{i+1}")
        processes.append(p)
        p.start()

    # Ensure all processes finish before the main script ends
    for p in processes:
        p.join()

    print("All processes have completed.")
```
