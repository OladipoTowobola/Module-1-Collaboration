```python
class Solution:
    def sort012(self, arr):
        low = 0
        mid = 0
        high = len(arr) - 1
        
        while mid <= high:
            if arr[mid] == 0:
                #Swap the current element with the element at low
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid+= 1
            elif arr[mid] == 1:
                # 1 is in the right place, just move forward
                    mid += 1 
            else:# arr[mid] == 2
                #Swap the current element with the element at high
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1 
                 #We don't increment mid here because the swapped element from 'high needs to be checked.   
        return arr
        # code here
        
```

```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) -1
        res = -1 # initialize with -1 in case k isn't found
        
        while  low <= high:
            mid = low + (high -low) // 2
            
            if arr[mid] ==k:
                res = mid # Potential answer found
                high= mid -1 # keep searching left for a smaller index
            elif arr[mid] < k:
                low = mid + 1 #k is the right half
            else:
                high =mid -1 #k is the left half
        return res
```
